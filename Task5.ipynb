Импортируем библиотеку pandas и сокращаем ее название до ps.

```python
import pandas as ps
```

Читаем файл 'kc_house_data.csv' в формате CSV и сохраняем его в переменной df в виде таблицы данных.

```python
df = ps.read_csv('kc_house_data.csv',encoding='UTF-8')
```

Используем метод info() для вывода информации о структуре таблицы данных, такой как количество строк и столбцов, названия столбцов и типы данных в каждом столбце.

```python
df.info()
```

### 1. Клиент хочет дом с видом на набережную(waterfront), как минимум с тремя ванными и с подвалом.

Выбираем строки, где значение в столбце 'waterfront' равно 1, значение в столбце 'bathrooms' больше или равно 3 и значение в столбце 'sqft_basement' не равно 0. И увидим сколько вариантов есть у клиента.

```python
df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)].shape[0]
```

Выведем все варианты таблицей

```python
df[(df['waterfront'] == 1) & (df['bathrooms'] >= 3) & (df['sqft_basement'] != 0)]
```

### 2. Клиент хочет дом

Либо с очень красивым видом(параметр 'view' - больше) из окна, либо с видом на набережную(waterfront). Таких квартир будет:

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5)].shape[0]
```

В очень хорошем состоянии(condition) и год постройки не меньше 1980 года. Таких квартир будет:

```python
df[(df['condition'] == 5) & (df['yr_built'] >= 1980)].shape[0]
```

Найдем минимальную и максимальную стоимости чтобы увидеть в каком ценовом диапазоне будут дома 

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price'].min()
```

```python
df[((df['view'] == 4) | (df['waterfront'] == 1)) & (df['condition'] == 5) & (df['yr_built'] >= 1980)]['price'].max()
```

### 3. Выберите дом клиенту. Клиент хочет дом

- без подвала
- с двумя этажами
- стоимостью до 150000. 

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]['condition']
```

- Какая оценка по состоянию у таких домов в среднем?

```python
df[(df['sqft_basement'] == 0) & (df['floors'] == 2) & (df['price'] < 150000)]['condition'].mean()
```
